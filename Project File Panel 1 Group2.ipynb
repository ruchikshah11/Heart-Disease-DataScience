## Feature description :

Quantitative data deals with numbers and things you can measure objectively: dimensions such as height, width, and length. Temperature and humidity. Prices. Area and volume.

Qualitative data deals with characteristics and descriptors that can't be easily measured, but can be observed subjectively — such as smells, tastes, textures, attractiveness, and color. 

From this definition's we figured out that in this data set all the values Quantitative.

Quantitative data has two sub types: 
1. Discrete : Discrete data is a count that can't be made more precise.Typically it involves integers.
2. Continuous : Continuous data, on the other hand, could be divided and reduced to finer and finer levels.

### Description :
             
1. `age:` in years 
 > Data Type : Discrete 
2. `sex:` 
        1 = male
        0 = female
 > Data Type : Discrete and categorical
3. `cp:` chest pain type 
      
      ***Angina may feel like pressure in the chest, jaw or arm. It frequently may occur with exercise or stress.***
 
      - Value **0** = typical angina : ***typical angina is the discomfort that is noted when the heart does not get enough blood or oxygen.***  
      - Value **1** = atypical angina : ***Atypical chest pain is defined as chest pain that does NOT have all three characteristics of chest pain that is typical of a heart attack.*** 
              1) chest pain or discomfort in the center of the chest behind the breastbone, 
              2) gets worse with exertion or stress, and
              3) gets better with rest or a medication called nitroglycerin 
      - Value **2** = non-anginal pain : ***Non-cardiac chest pain is the term that is used to describe pain in the chest that is not caused by heart disease or a heart attack. Causes include muscle or bone problems, lung conditions or diseases, stomach problems, stress, anxiety, and depression.***

     > Data Type : Discrete and categorical

4. `trestbps:` Resting blood pressure(Optimal blood pressure) measured in mmHg(millimeters of mercury) on admission to the hospital
    
     ***A normal reading would be any blood pressure below 120 mm Hg and above 90 mm Hg in an adult.***
  > Data Type : Discrete   
           

5. `chol:` Serum cholesterol in Milligrams Per Deciliter(mg/dl)
      
      ***The total serum cholesterol level is the amount of cholesterol in the blood.***
  > Data Type : Discrete

          0 = false
  > Data Type : Discrete and categorical

     - Value 0: showing probable or definite left ventricular hypertrophy by Estes' criteria 
          
          ***Left ventricular hypertrophy, or LVH, is a term for a heart's left pumping chamber that has thickened and may not be pumping efficiently.***
     - Value 1: normal
     - Value 2: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV(millivolts))
         
         ***ST elevation refers to a finding on an electrocardiogram wherein the trace in the ST segment is abnormally high above the baseline.Inverted T wave is considered abnormal if inversion is deeper than 1.0 mm.***
  
  > Data Type : Discrete and categorical

8. `thalach:` Maximum heart rate achieved
  > Data Type : Discrete

### Hypothesis Tests for Two Proportions

- A two-sample test of proportions is performed to assess if the population proportion of some traits differs between two subgroups. 

- Here, we are going to test if the population proportion of male with heart disease is different from the population proportion of female with heart disease.

**Null Hypothesis:** $(H_0): p_1 - p_2 = 0$

**Alternate Hypothesis:** $(H_a): p_1 - p_2 \ne 0$

 - $p_1$ is proportion of male with heart disease;  
 
 - $p_2$ is proportion of female with heart disease.

**Level of significance: $\alpha = 0.05$**

**Formula:**

\begin{equation}
z=\frac{p_1-p_2}{SE};
\end{equation}

The standard error for two population proportion is calculated with the formula below:

\begin{equation}
SE=\sqrt{p\left( 1-p \right)\left( \frac{1}{{{n}_{1}}}+\frac{1}{{{n}_{2}}} \right)} 
\end{equation}

where, $p$ is the total population proportion in the sample with heart disease.

$n_1$ and $n_2$ are the total numbers of the male and female in the sample.

```python
pd.set_option('precision', 4)
df_cross1 = pd.crosstab(df_hr.target, df_hr.sex, normalize=True, margins=True)
df_cross = pd.crosstab(df_hr.target, df_hr.sex, margins=True)
```

```python
# Total proportion of the patients with heart disease
p_null = df_cross1['All'].yes

# sample size
n1 = df_cross['male'].All
n2 = df_cross['female'].All

# level of significance
alpha = 0.05

# sample proportion
p1 = df_cross1['male'].yes
p2 = df_cross1['female'].yes

# Standard Error
se = np.sqrt(p_null * (1 - p_null) / (1/n1 + 1/n2))

# Test statistic
z_cal = (p1 - p2) / se

z_tab = norm.ppf(1-alpha/2)

df_ztest = pd.DataFrame([{"Test Statistic": z_cal, 
                         "Critical Value": z_tab}],
                        index=["Sample Data"]
                       )
df_ztest.style.background_gradient(cmap='coolwarm')
```

Since our test statistic value is less than the critical value, we have good evidence not to reject the null-hypothesis at the  0.05  significance level.

Hence, we can infer that there is no significant difference in the proportion of male and female with heart disease.
